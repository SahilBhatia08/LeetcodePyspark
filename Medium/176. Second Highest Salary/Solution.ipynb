```python
# Libraries
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
```

```python
# Define Schema
Schema = StructType([
    StructField("id", IntegerType()),
    StructField("salary", IntegerType())
])
```

```python
# DataFrame Example 1
data = [(1,100),(2,200), (3,300)]

employee_df = spark.createDataFrame(data, Schema)
employee_df.show()
```

```python
# Solution
def nth_highest_salary(df, n):
    # Defining Window Spec
    windowSpec = Window.orderBy(col("salary").desc())
    # Dropping Duplicate and Applying Dense Rank to find nth highest salary
    rankedDf = df.select(col("salary")).distinct().select(
        col("salary"),
        dense_rank().over(windowSpec).alias("rank")
    )
    #filter for nth highest salary
    filteredDf = rankedDf.filter(col("rank") == n).select(col("salary").alias("nthHighestSalary"))

    return filteredDf.select(max(col("nthHighestSalary")).alias("nthHighestSalary"))
```

```python
#Result 1
nth_highest_salary(employee_df, 2).show()
```

```python
#DataFrame Example 2
data = [(1,100),]

employeeDf2 = spark.createDataFrame(data, Schema)
employeeDf2.show()
```

```python
nth_highest_salary(employeeDf2, 2).show()
```
