```python
# Libraries
from pyspark.sql.functions import *
from pyspark.sql.types import *
from pyspark.sql.window import Window
```

```python
# Define Schema
Schema = StructType([
    StructField("id", IntegerType(), False),
    StructField("num", StringType())
])
```

```python
# DataFrame Example 1
data = [(1,1),(2,1), (3,1), (4,2), (5,1), (6,2), (7,2)]

LogDf = spark.createDataFrame(data, Schema)
LogDf.show()
```

```python
def ThreeConsecutiveNumbers(df):
    #Defining Window Spec
    windowSpec = Window.orderBy(col("id").asc())

    #Using Lead and Lag to get the next and previous values
    NextDf = df.withColumn("next", lead("num", 1, "0").over(windowSpec))
    PrevDf = NextDf.withColumn("prev", lag("num", 1, "0").over(windowSpec))

    #Finding Consecutive Number
    ConsecutiveDf = PrevDf.filter(
        (col("num") == col("next")) & 
        (col("num") == col("prev"))
        )\
        .select(
            col("num")
            .alias("ConsecutiveNums")
        )\
        .distinct()
    
    return ConsecutiveDf
```

```python
ThreeConsecutiveNumbers(LogDf).show()
```
